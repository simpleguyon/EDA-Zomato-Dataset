```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
##Check dtypes of all columns
final_df.dtypes
```

```python
 country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#Top 3 country using zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Obversation: Zomato's maximum records are from India following by USA and UK. 

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#Obversation
1. When the rating is between 4.5 to 4.9 ---> Excellent
2. When the rating is between 4.0 to 4.4 ---> Very good
3. When the rating is between 3.5 to 3.9 ---> Good
4. When the rating is between 2.5 to 3.4 ---> Average
5. When the rating is between 1.8 to 2.4 ---> Poor
6. When the rating is 0 ---> No Rating Given

```python
plt.rcParams['figure.figsize'] = (11,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['Black','Red','Orange','yellow','Green','Green'],)
plt.legend(loc='upper right')
```

Obversation:
   1. Not rated country is highest
   2. Max no of. ratings between 2.5 to 3.4

```python
#Country Plot
sns.countplot(x='Rating color',data=ratings,palette=['Black','Red','Orange','yellow','Green','Green'])
```
